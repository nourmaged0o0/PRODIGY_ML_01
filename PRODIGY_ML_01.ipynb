# Import trainning data

```python
import pandas as pd
data = pd.read_csv('/content/train.csv')
data.head()
```

# Use necessary columns

```python
data = data[['BedroomAbvGr', 'FullBath', 'HalfBath', 'GrLivArea', 'LotArea', 'SalePrice']]
```

```python
data.head()
```

```python
data.shape
```

# Check if there is a outlayers

```python
data['BedroomAbvGr'].value_counts()
data['FullBath'].value_counts()
```

# Split training data to features and labels

```python
x = data.iloc[:,0:5]
y = data.iloc[:,5:6]
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaleX= MinMaxScaler()
scaleY= MinMaxScaler()
x = scaleX.fit_transform(x)
y = scaleY.fit_transform(y)
```

# Build the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=80)
reg = LinearRegression()
reg.fit(x_train,y_train)

score = cross_val_score(reg,x,y,cv=5)
score.mean()
```

# Import testing data

```python
data2 = pd.read_csv('/content/test.csv')
data2.head()
```

# Use necessary columns

```python
data2=data2[['BedroomAbvGr', 'FullBath', 'HalfBath', 'GrLivArea', 'LotArea']]
data2.head()
```

# Scale the features of TestingData

```python
x2=data2.copy()
x2 = scaleX.fit_transform(x2)
```

# Use the model to predict the prices of testing data

```python
y2 = reg.predict(x2)
```

# Inverse scaling from predicted prices

```python
answer = scaleY.inverse_transform(y2)
answer
```

# Add price column to new data

```python
data2['SalePrice']=answer
data2.head()
```

# Export new TestingData with predicted prices column

```python
data2.to_csv('predicted_prices.csv',index=False)
```
